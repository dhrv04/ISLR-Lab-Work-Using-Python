## High Leverage Points and The Hat Matrix

    High Leverage points are the points that have an unusual value of x_i, these values can have an sizeable 
    impact on the shape of our regression line, hence it is important to identify theae high leverage pointa.

    The impact of high leverage points can be shown as:

<img src="https://i.stack.imgur.com/XCaiG.png">

    The hat matrix is a projection matrix that transforms the values of the response variable i.e y into the
    linear combinations of predictors X's that our model is regressing on i.e y_cap, as the predicted value of 
    y_cap is different from the actual value of y, the difference is equal to residual

<img src="https://i.stack.imgur.com/r6umZ.png">

    The coefficients of our regression model are considered as linear combinations of the predictor values.
    Hence, we have:

\begin{equation} \hat{H}.y = \hat{y} \end{equation}

    The value of Hat Matrix is calculated as:

\begin{equation} H = X(X^{T}X)^{-1}X^{T} \end{equation}

    The Leverage score of particular observation of our training data can be found in the diagonal of the hat matrix
    and the entry(observation) will be considered to have a high leverage if value of the leverage score in the 
    diagonal is greater than:

<a href="https://www.codecogs.com/eqnedit.php?latex=2\sum&space;^{n}_{i=1}&space;h_{ii}/n" target="_blank"><img src="https://latex.codecogs.com/gif.latex?2\sum&space;^{n}_{i=1}&space;h_{ii}/n" title="2\sum ^{n}_{i=1} h_{ii}/n" /></a>

### Cook's Distance

Reference = https://www.statisticshowto.com/cooks-distance/#:~:text=In%20other%20words%2C%20it's%20a,26%20as%20a%20potential%20outlier.

```python
model = smf.ols(formula=model_formula, data=df)

model_fit = model.fit()

# Fitted values are also called as predicted values
model_fitted_y = model_fit.fittedvalues

model_residuals = model_fit.resid

# Now in order to find outliers we need to studenentize or normalize the residuals.
# Studentizing the residual value is done by dividing the residual by its standard error.
model_norm_residuals = model_fit.get_influence().resid_studentized_internal

# Square-root of residuals (If the absolute value is greater than 3 then the given residual is an outlier)
model_norm_residuals_abs_sqrt = np.sqrt(np.abs(model_norm_residuals))

# Absolute value of residuals
model_residuals_abs = np.abs(model_residuals)


# We studentized our residuals to show outliers, but we still need to get know high leverage points
# High Leverage Points are the points which have usual values if x_i i.e the value of the predictor
# In order to quantify leverage statistic we compute leverage statistic.
model_leverage = model_fit.get_influence().hat_matrix_diag

model_cooks = model_fit.get_influence().cooks_distance[0]
```

```python
model_cooks.shape
```

```python
model_cooks
```

```python
# Plotting residual values against predicted values

lm_plot = plt.figure(1)
lm_plot.set_figheight(8)
lm_plot.set_figwidth(12)

lm_plot.axes[0] = sns.residplot(model_fitted_y, 'mpg', data=df, lowess=True, scatter_kws={'alpha':0.5}, line_kws=
                               {'color':'red', 'alpha':0.8, "lw":1})

lm_plot.axes[0].set_title("Residual v/s Fitted")
lm_plot.axes[0].set_xlabel("Fitted Values")
lm_plot.axes[0].set_ylabel('Residual')

# Annotating the Outliers
abs_resid = model_residuals_abs.sort_values(ascending=False)
abs_resid_top_3 = abs_resid[:3]

for index in abs_resid_top_3.index:
    
    lm_plot.axes[0].annotate(s=index, xy=(model_fitted_y[index], model_residuals[index]))
```

    The above plot of residual values versus the predicted values clearly shows skewness and that the relation
    between the predictor and response variable is not linear also a few outliers can be detected

```python
abs_resid_top_3.index
```

    (4) Residuals against leverages

    Plotting residuals against leverages tell us about the various possible leverage points in our data, Here we 
    plot leverages againt the normalized residual values in order to check whether either of the data points in 
    our dataframe lies above the cook's boundary.

```python
plot_4 = plt.figure(4)
plot_4.set_figheight(8)
plot_4.set_figwidth(12)

plt.scatter(model_leverage, model_norm_residuals)
# lowess: stands for locally weighted linear regression which smoothens the curve and let's us see the clarity of the curve
sns.regplot(model_leverage, model_norm_residuals, lowess=True, line_kws={'color': 'r'})  

plot_4.axes[0].set_xlim(0.00,0.20)
plot_4.axes[0].set_ylim(-3, 5)
plot_4.axes[0].set_title("Studentized Residual v/s Leverage")
plot_4.axes[0].set_xlabel("Leverages")
plot_4.axes[0].set_ylabel("Normalized Residuals")

leverage_top_3 = np.flip(np.argsort(model_cooks),0)[:3]

# Adding Annotation of highest cook's distance values on the graph
for i in leverage_top_3:
    
    plot_4.axes[0].annotate(i, xy=(model_leverage[i], model_norm_residuals[i]))
```
